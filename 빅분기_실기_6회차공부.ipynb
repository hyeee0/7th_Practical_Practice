## [1] 대응(쌍체)표본 T검정

주어진 데이터는 고혈압 환자 치료 전후의 혈압이다.<br>
해당 치료가 효과가 있는지 대응(쌍체)표본 t-검정을 진행하시오


*   귀무가설(H0):  μ >= 0<br>
*   대립가설(H1):  μ < 0<br>
*   μ= (치료 후 혈압 - 치료 전 혈압)의 평균<br>
*   유의수준: 0.05<br>

1. μ의 표본평균은?(소수 둘째자리까지 반올림)
2. 검정통계량 값은?(소수 넷째자리까지 반올림)
3. p-값은?(소수 넷째자리까지 반올림)
4. 가설검정의 결과는? (유의수준 5%)

```python
import pandas as pd
from scipy import stats #정규

df = pd.read_csv('/content/high_blood_pressure.csv')
df
# μ= (치료 후 혈압 - 치료 전 혈압)의 평균

df['diff'] = df['bp_post'] - df['bp_pre']
print(round(df['diff'].mean(),2)) # μ의 표본평균은?(소수 둘째자리까지 반올림)

st, pv = stats.ttest_rel(df['bp_post'], df['bp_pre'], alternative='less') # 대응 표본 t검정 -> ttest_rel()
#alternative='less' -> 가설에서 작은지가 궁금하니깐
print(round(st, 4)) # 검정통계량 값은?(소수 넷째자리까지 반올림)
print(round(pv, 4)) # p-값은?(소수 넷째자리까지 반올림)

# p_value가 0.0016 < 0.05 이므로 귀무가설 기각, 대립가설 채택
```

대응 표본 t-검정
- 동일한 모집단으로부터 추출된 두 표본 집단이 대상<br>

<scipy - ttest_rel()>
- 대응 표본 t검정 실시할 때 사용
- 검정에 실시하는 두 변수 차례대로 지정

## [2] 독립표본 T검정
어떤 특정 약물을 복용한 사람들의 평균 체온이 복용하지 않은 사람들의 평균 체온과 유의미하게 다른지 검정해보려고 합니다.


*   약물을 복용한 그룹과 복용하지 않은 그룹의 체온 데이터가 각각 주어져 있다고 가정합니다.
*   각 그룹의 체온은 정규분포를 따른다고 가정합니다.

검정통계량, p-value, 검정결과를 출력하시오<br>




```python
from scipy import stats

# 가설 설정
# H0: 약물을 복용한 그룹과 복용하지 않은 그룹의 평균 체온은 유의미한 차이가 없다. (귀무가설)
# H1: 약물을 복용한 그룹과 복용하지 않은 그룹의 평균 체온은 유의미한 차이가 있다. (대립가설)

# 데이터 수집
group1 = [36.8, 36.7, 37.1, 36.9, 37.2, 36.8, 36.9, 37.1, 36.7, 37.1]
group2 = [36.5, 36.6, 36.3, 36.6, 36.9, 36.7, 36.7, 36.8, 36.5, 36.7]

st, pv = stats.ttest_ind(group1, group2)
print('검정통계량', st)
print('p-value', pv)

alpha = 0.05

if pv < alpha: # p-value가 0.05보다 작으니깐 대립가설 채택
    print('귀무가설을 기각합니다. 약물을 복용한 그룹과 복용하지 않은 그룹의 평균 체온은 유의미한 차이가 있습니다.')
else:
    print('귀무가설을 채택합니다. 약물을 복용한 그룹과 복용하지 않은 그룹의 평균체온은 유의미한 차이가 없습니다.')
```

독립 2 표본 t-검점 <br>
- 독립된 두 표본집단이 대상
- 등분산 여부에 따라 검정통계량 계산식이 다름
- 보통 t-test하면 이 경우가 제일 많다<br>

<scipy - ttest_ind()>
- 독립 2 표본 t검정 실시할 때 사용
- 검정에 실시하는 두 변수 차례대로 지정

## [3] 단일표본 T검정
다음은 22명의 학생들이 국어시험에서 받은 점수이다. 학생들의 평균이 75보다 크다고 할 수 있는가?

*   귀무가설(H0): 모평균은 mu와 같다. (μ = mu), 학생들의 평균은 75이다
*   대립가설(H1): 모평균은 mu보다 크다. (μ > mu), 학생들의 평균은 75보다 크다

<가정>
*   모집단은 정규분포를 따른다.
*   표본의 크기가 충분히 크다.

검정통계량, p-value, 검정결과를 출력하시오





```python
from scipy.stats import ttest_1samp

scores = [75, 80, 68, 72, 77, 82, 81, 79, 70, 74, 76, 78, 81, 73, 81, 78, 75, 72, 74, 79, 78, 79]

mu = 75
alpha = 0.05

st, pv = stats.ttest_1samp(scores, mu, alternative='greater') #alternative='greater' -> 가설에서 큰지가 궁금하니깐 greater을 사용
print('검정통계량', st)
print('p-value', pv)

if pv < alpha:
    print('귀무가설을 기각합니다. 모평균은 75보다 큽니다.')
else:
    print('귀무가설을 채택합니다. 모평균은 75보다 크지 않습니다.')
```

단일 표본 t 검정(One Sample t-test)<br>
- 단일 모집단에서 추출된 하나의 표본이 대상<br>
- 원래 t검정은 두 집단 간의 평균 비교하는 건데 샘플이 하나 <br>-> 하나는 모평균임 <br>->> 모평균과 표본평균의 차이 검정<br>

<scipy - ttest_1samp()>

## [4] t3-anova

세 가지 다른 교육 방법(A, B, C)을 사용하여 수험생들의 시험 성적을 개선시키는 효과를 평가하고자 한다. 30명의 학생들을 무작위로 세 그룹으로 배정하여 교육을 실시하였고, 시험을 보고 성적을 측정하였습니다. 다음은 각 그룹의 학생들의 성적 데이터입니다.



*   귀무가설(H0): 세 그룹(A, B, C) 간의 평균 성적 차이가 없다.
*   대립가설(H1 또는 Ha): 세 그룹(A, B, C) 간의 평균 성적 차이가 있다.

일원배치법을 수행하여 그룹 간의 평균 성적 차이가 있는지 검정하세요

1. f값 (소수 둘째자리)
2. p값 (소수 여섯째자리)
3. 검정결과 출력


```python
# 각 그룹의 데이터
groupA = [85, 92, 78, 88, 83, 90, 76, 84, 92, 87]
groupB = [79, 69, 84, 78, 79, 83, 79, 81, 86, 88]
groupC = [75, 68, 74, 65, 77, 72, 70, 73, 78, 75]
```

```python
import scipy.stats as stats

# 일원배치법 -> anova
f_value, p_value = stats.f_oneway(groupA, groupB, groupC)
print(round(f_value, 2))
print(format(p_value, '.6f'))
```

## [5] 정규성 검정

12명의 수험생이 빅데이터 분석기사 시험에서 받은 점수이다. Shapiro-Wilk 검정을 사용하여 데이터가 정규 분포를 따르는지 검증하시오
<br> 귀무가설:데이터는 정규분포를 따른다
<br> 대립가설 : 데이터는 정규분포를 따르지 않는다

```python
from scipy import stats

data = [75, 83, 81, 92, 68, 77, 78, 80, 85, 95, 79, 89]

# Shapiro-Wilk 검정수행
statistic, p_value = stats.shapiro(data)

print("Shapiro-Wilk 검정 통계량 : ", statistic)
print("p-value : ", p_value)

alpha = 0.05
if p_value > alpha:
    print("귀무 가설을 기각할 수 없다. 데이터는 정규 분포를 따름")
else:
    print('귀무가설을 기각한다. 데이터는 정규분포를 따르지 않음')
```

## [6] 회귀모형(상관계수)
iris에서 Sepal Length와 Sepal Width의 상관계수 계산하고 소수 둘째자리까지 출력


```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# 상관계수 계산
correlation = df.corr()
# sepal length와 sepal width의 상관계수 계산이니깐 두개의 컬럼을 loc를 통해 뺴야한다
result = correlation.loc['sepal length (cm)', 'sepal width (cm)']

print(round(result, 2)) # 반올림해서 소수점 둘째짜리까지 표현
```

#### 수치형 변수 스케일링
- 중요 : train 데이터만 fit해준다 <br> -> 즉, train에는 tit_transform, test에는 transform

```python
from sklearn.preprocessing import StandardScaler

con_df = train.select_dtypes(include=['int64', 'float64']).copy() # 수치형 변수를 복사
con_cols = con_df.columns

# 스케일링 전 확인
display('train 스케일링 전', train.head())
display('test 스케일링 전', test.head())

# 스케일링을 해주자~

scaler = StandardScaler()

train[con_cols] = scaler.fit_transform(train[con_cols])
test[con_cols] = scaler.transform(test[con_cols])

# 스테일링 후 확인

display('train 스케일링 후', train.head())
display('test 스케일링 후', test.head())
```

- 선형회귀는 원-핫 인코딩을 적용
- 선형회귀에 라벨 인코더를 적용하면 문제가 발생한다 <br> ->
회귀 계수 연산시, 라벨 인코딩 된 후 숫자의 크고 작음이 모델에 부정적 영향을 끼칠 수도 있다.

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(train, y, test_size = 0.2, random_state=2023)
print(X_train.shape, X_val.shape, y_train.shape, y_val.shape)
```

### [4] 회귀 모델링 및 비교
- 회귀에 많이 쓰는 랜덤포레스트, xgboost, lgbm 비교
- RMSE 값이 작을수록 모델의 성능이 좋다고 평가<br> 평균 제곱근 오차는 회귀 예측 모델에 대한 두 개의 주요 성과 지표 중 하나, 완벽한 예측 모델은 평균 제곱근 오차 값이 0이다

```python
from sklearn.ensemble import RandomForestRegressor # 복원추출
from xgboost import XGBRegressor # 비복원추출
from lightgbm import LGBMRegressor # 비복원추출
```

```python
# rmse
from sklearn.metrics import mean_squared_error
import numpy as np

def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
```

#### 랜덤포레스트

```python
rf = RandomForestRegressor(random_state=2023)
rf.fit(X_train, y_train)
rf.pred = rf.predict(X_val) # 검증용
rmse(y_val, rf.pred)
```

#### XGBRegressor

```python
xgb = XGBRegressor(random_state=2023)
xgb.fit(X_train, y_train)
xgb.pred = xgb.predict(X_val)
rmse(y_val, xgb.pred)
```

#### LGBMRegressor

```python
lgbm = LGBMRegressor(random_state=2023)
lgbm.fit(X_train, y_train)
lgbm.pred = lgbm.predict(X_val)
rmse(y_val, lgbm.pred)
```
